**Descriptive : (summary statistics)**

when we perform any methamatical operation on given data and as a result describe some characteristics of that data. applied to both sample as well as population.
examples: mean,mode,median,std,variance,covariance

**Inferential**

When we select small sample from large population and perform analysis on it to draw some conclusion for the entire population.

**BASIC TERMS**

- a.	sample (on which actually analysis perform)
1.	should be random
2.	representive
- b.	population (the entire data pool from which sample is drawn)
- c.	statistics (the metrics performed on sample)
- d.	parameter (the metrics performed on population)
