```python
import numpy as np
import pandas as pd

company = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CLIENTS/7. Dammy/Pandas/Fortune 500 2017 - Fortune 500.csv')

# Set display options to show maximum columns
pd.set_option('display.max_columns', None)

company.head()
```

#General understanding of group bt function

```python
# Group by function will create groups dataframes equal to the unique value present in that column
sectors = company.groupby('Sector')
sectors
```

```python
# To see all the groups and their corresponding rows
sectors.value_counts()
```

```python
# To see the number of instances that particular value comes up in our data
# It is like value counts function on dataframe
sectors.size().sort_values(ascending=False)
```

```python
# number of groups
len(sectors)
```

```python
# To select a particular group
sectors.get_group('Financials')
```

```python
sectors.groups
# It returns the dictionary which will provide all the companies indexes of that corresponding sector
```

##Q#5.Can you provide a breakdown of profits(average) for each industry, grouping the data by industry and displaying the mean profits generated by companies within each industry?

```python
# Set display options to show maximum columns
pd.set_option('display.max_rows', None)

industries = company.groupby('Industry')
industries['Profits'].mean().sort_values(ascending=False)
```
