#How many matches are played in each season

- As the id and match id represent the same information so we can join these two datasets together based on this common key feature.

```python
merged_df = delivery.merge(matches,left_on='match_id',right_on='id')
merged_df.head()
```

```python
merged_df.groupby('season').id.nunique()
```

# The highest run getter of each season?

- We have to merge these two files,but it is possible only when there is some link present between two data sets.like in this case match_id

```python
merged_df.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index().drop_duplicates(subset = 'season',keep = 'first').sort_values('season')
```
