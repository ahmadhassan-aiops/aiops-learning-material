#Give me the visualization for 5 most frequent sectors!

```python
# first we have to fetch the respective data
data['Sector'].value_counts()
```

##1.Line plot

```python
# by default it will give us line plot
data['Sector'].value_counts().plot()
```

##2.Bar plot

```python
# by default it will generate vertical bar plot
data['Sector'].value_counts().head().plot(kind='bar')
```

```python
# it will generate horizontal bar plot
data['Sector'].value_counts().head().plot(kind='barh')
```

```python
data['Sector'].value_counts().plot(kind='barh',figsize=(8,4))
```

##3.Pie plot

```python
data['Sector'].value_counts().head().plot(kind='pie')
```

##4.Histogram

```python
data['Revenues'].plot(kind='hist')
```
