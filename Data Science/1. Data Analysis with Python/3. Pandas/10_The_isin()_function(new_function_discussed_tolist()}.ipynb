##Q#1.Which sector has the highest average revenue among companies that have more than 50,000 employees and have experienced positive profit change?



- Any non-numeric values (like strings or missing values) are converted to NaN, preventing errors during filtering

```python
# Convert Employees to numbers, coerce invalid values to NaN
company['Employees'] = pd.to_numeric(company['Employees'], errors='coerce')
company['Prftchange'] = pd.to_numeric(company['Prftchange'], errors='coerce')

filtered_df = company[(company['Employees'] > 50000) & (company['Prftchange'] > 0)]

filtered_df.groupby('Sector')['Revenues'].mean().sort_values(ascending=False)
```

##Q#2.Which industry has the highest total revenue, but only for companies where profit margins (profits/revenues) are greater than 5% and they have more than 100,000 employees?



```python
company['Employees'] = pd.to_numeric(company['Employees'], errors='coerce')
company['Profits'] = pd.to_numeric(company['Profits'], errors='coerce')
company['Revenues'] = pd.to_numeric(company['Revenues'], errors='coerce')

filtered_df = company[(company['Profits'] / company['Revenues'] > 0.05) & (company['Employees'] > 100000)]

filtered_df.groupby('Industry')['Revenues'].sum().sort_values(ascending=False)
```
