#a. Heatmaps

- Heatmaps are a method of representing data graphically where values are depicted by color, making it easy to visualize complex data and understand it at a glance.
- It is developed b/w two categorical data,also can be developed for numerical data but it has to be discrete


```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=sns.load_dataset("flights")
data.head()
```

```python
data=sns.load_dataset("tips")
data.info()
```

```python
# to remove the numeric features from tips data
data=data.select_dtypes(include=['number'])
data.head()
```

```python
sns.heatmap(data)
```

```python
flight_data=data.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
flight_data
```

```python
sns.heatmap(flight_data)
```

**CUSTOMIZATION**

```python
sns.heatmap(flight_data,cbar=True)
```

```python
sns.heatmap(flight_data,linewidths=1)
```

```python
sns.heatmap(flight_data,linewidths=1,annot=True,fmt='d')
#d specifies integers values which is needed here
```

```python
sns.heatmap(flight_data,linewidths=1,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(flight_data,linewidths=1,annot=True,fmt='d',cmap='summer')
```

#cluster maps
- Clustering is basically grouping data based on relationships among the variables in the data. Clustering algorithms help in getting structured data in unsupervised learning.
- Simple we can say that it clustered similar rows and columns together


```python
sns.clustermap(flight_data,linewidths=1,annot=True,fmt='d')
```

**Clustring happens on the basis of machine learning distances,which is stored here in the "metric" attribute which by defualt is eucledean**

```python
sns.clustermap(flight_data,linewidths=1,annot=True,fmt='d',metric='correlation')
```

**score leveling**

```python
sns.clustermap(flight_data,metric='correlation',annot=True,z_score=0)
```

**annotation**

```python
sns.clustermap(flight_data,metric='correlation',fmt='d',annot=True)
```

**TO DECIDE WHETHER CLUSTIRING IS BASED ON ROWS OR COLUMNS**

```python
sns.clustermap(flight_data,metric='correlation',col_cluster=False)
```
