```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

#plotting with categorical data
**it is done by using "catplot()" function which is a figure level function**

**1. Categorical Scatterplots**

- a.Stripplots
- b.Swarmplots

- **x axis** == categorical data
- **y-axis** == numerical data


```python
data=sns.load_dataset("tips")
data.head()
```

```python
data.shape
```

#**a. stripplots(on tip and day column)**
the by default kind value of catplot is strip thus does'nt need to be mentioned
- Two values can be interpretted from striplot:
1. Density (occurance of specific category)
2. Range of specofic category

- **The limitation for these graphs is that size of data should be small**

```python
sns.catplot(x='smoker',y='total_bill',data=data)
```

**JITTER CONCEPT**

A jitter plot is a variant of the strip plot with a better view of overlapping data points, used to visualise the distribution of many individual one dimensional values. by default jitter values work just fine in almost all cases


```python
sns.catplot(x='time',y='tip',data=data,jitter=0.1)
```

```python
sns.catplot(x='time',y='tip',data=data,jitter=0)
```

```python
# plt.style.use('ggplot')
# plt.figure(figsize=(8,5))
sns.stripplot(x='time',y='tip',data=data,jitter=0.2)
```

#**2. swarm plots (distribution of data is well explained)**

```python
sns.catplot(x='time',y='tip',data=data,kind='swarm')
```

#SEMANTIC PARAMETERS
- I.E SEX AS HUE PARAMETER.
- it is the only parameter that can be aplied here

```python
sns.catplot(x='time',y='tip',data=data,kind='swarm',hue='sex')
```

**instead of provind kind argument to catplot we can build scatter plot directly**

```python
sns.swarmplot(x='time',y='tip',data=data,hue='sex')
```
