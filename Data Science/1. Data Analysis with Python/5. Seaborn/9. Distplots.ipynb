A Distplot or distribution plot, depicts the variation in the data distribution. Seaborn Distplot represents the overall distribution of continuous data/numerical data variables. The Seaborn module along with the Matplotlib module is used to depict the distplot with different variations in it.
**combine 3 things in one graph**
- a.	histogram
- b.	kemel density estimation (PDF)
- c.  ECDF (Empirical Cumulative Distribution Function)
- c.	rug plot


```python
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=sns.load_dataset("titanic")
data.head()
```

```python
data.shape
```

```python
data.isna().sum()/data.shape[0]*100
```

**There are some missing values in age column we will replace them with mean age values**

```python
data['age'].mean()
```

```python
data['age'] = data['age'].fillna(data['age'].mean())
data.isna().sum()
```

**By default histogram and PDF is shown,but for rugplot we have to add its value**

```python
sns.distplot(data['age'],kde=True)
```

```python
sns.displot(data['age'])
```

```python
sns.histplot(data['age'])
```

```python
sns.displot(data['age'],kind='kde')
```

```python
sns.kdeplot(data['age'])
```

```python
sns.displot(data['age'],kind='ecdf')
```

```python
sns.ecdfplot(data['age'])
```

**ONLY HISTOGRAM**

**1. same like sns.histplot**


```python
sns.distplot(data['age'],kde=False)
```

**bins can be manually set**

```python
sns.distplot(data['age'],kde=False,bins=10)
```

**ONLY KDE(PDF)**

**1. same like sns.displot with providing kind=kde**

```python
sns.distplot(data['age'],hist=False)
```

**only rug plot**

A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualise the distribution of the data.As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot

```python
data
```

```python
sns.distplot(data['age'],hist=False,kde=False,rug=True)
```

```python
sns.rugplot(data['age'])
```

**HOW TO USE DISTPLOT?**

I.E IF WE WANT TO FILTERED THE LIST OF PEOPLE WHO SURVIVED AND THEN PLOT THEM WITH RESPECT TO AGES.

```python
sns.distplot(data[data['survived']==1]['age'],hist=False)
sns.distplot(data[data['survived']==0]['age'],hist=False)
# brown line....dead
# blue line......living
```
