**1.box plot with numerical data(univariate analysis)**

##For explanation of findings in box plot

- Normal range of data
- IQR of data(where the 50% of data is concentrated)
- The nature of outliers (if present)

```python
data=sns.load_dataset("tips")
data.head()
```

```python
data.shape
```

```python
sns.boxplot(x="tip",data=data)
# the tips are numerical data only(an example of univariate analysis)
```

```python
sns.kdeplot(data["tip"])
```

```python
data['tip'].max()
```

```python
np.percentile(data["tip"],75)
```

```python
np.percentile(data["tip"],25)
```

```python
3.5625-2
```

```python
# lower limit for oitlier detection
2-(1.5*1.5625)
```

```python
# higher limit for oitlier detection
3.5625+(1.5*1.5625)
```

**box plot with categorical data + numerical data (bivariate analysis)**

```python
sns.boxplot(x='sex',y='tip',data=data)
```

**BY USING CATPLOT FUNCTION**

```python
sns.catplot(x='time',y='total_bill',data=data,kind='box')
```

**with hue parameter**

```python
sns.catplot(x='time',y='total_bill',data=data,kind='box',hue='sex')
```
