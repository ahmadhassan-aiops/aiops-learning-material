**a. BARPLOT**

- I.E WE WANTED TO CALCULATE THE MEAN TOTAL BILL OF SMOKER AND NON-SMOKERS( PLOT B/W NUMERICAL AND CATEGORICAL DATA)
- the **black line is error bar**
- An error bar is a line through a point on a graph, parallel to one of the axes, which represents the uncertainty or variation of the corresponding coordinate of the point.
- They are usually displayed as error bars on a graph. A **95% confidence limit** means that there is only a 5% chance that the true value is NOT included within the span of the error bar. This is a way of visualizing uncertainty in summary points plotted in a graph

```python
# here the bar plot is not correct (not showing right info)
plt.bar(data['size'],data['total_bill'])
```

```python
# Calculate the mean total_bill for each size
mean_total_bill = data.groupby('size')['total_bill'].mean()

# Use plt.bar() with aggregated data
plt.bar(mean_total_bill.index, mean_total_bill.values)
plt.xlabel('Size')
plt.ylabel('Mean Total Bill')
plt.title('Mean Total Bill by Size')
plt.show()
```

**with hue parameter**

```python
sns.barplot(x='smoker',y='tip',data=data,hue='sex')
```

- **the default behavior of barplots is to calculate the mean value of numerical data but can be changed by calling the estimator attribute**
- **estimator=np.median/var/std/sum**

```python
sns.barplot(x='smoker',y='total_bill',data=data,hue='sex',estimator=np.sum)
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,hue='sex',estimator=np.std,kind='bar')
```

**b. countplots**

Show the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. **The basic API and options are identical to those for barplot()** , so you can compare counts across nested variables. i.e if we want to calculate # of male and female from our data set


```python
data['size'].value_counts()
```

```python
sns.countplot(x='size',data=data)
```

**can be used with hue parameter**

```python
sns.countplot(x='day',data=data,hue='smoker')
```
