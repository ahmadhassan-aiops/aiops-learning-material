#Seaborn
Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive. It is built on the top of matplotlib library and also closely integrated into the data structures from pandas.

#SCATTER PLOTS

to draw scatter plot in seaborn:
- 1.scatterplot()....axis level function
- 2.relpot()......figure level function

- When we want to see a relationship between two numerical variables

```python
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import seaborn as sns
sns.get_dataset_names()
```

**in built data set of seaborn library**

```python
data=sns.load_dataset("tips")
```

```python
data.head()
```

**scatter plot b/w total_bill and tip**

to establish a relationship between them,for this we use relplot function becauseby using this we can generate both line and scatter plots....in case of scatterplots there is no need to mention the kind value which is scatter plot by default.the linear relationship is established between total_bill and tip

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)
```

#SEMANTIC PARAMETERS

**1. HUE PARAMETER (categorical data is passed)**

by using i.e smoker parameter we can conclude that non smokers follow linear
relation while smokers doesnot

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

**2. STYLE (categorical data is passed)**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

**3.SIZE (numerical data (discrete)is passed)**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

**customized sizes**

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(60,80),data=data)
```

**USING THE SCATTER PLOT**

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
plt.xlabel('the total bill')
```
