A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the probability density of each variable.

image.png

```python
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
data.head()
```

**1. for only numerical data (univariate analysis)**

```python
sns.violinplot(x='tip',data=data)
```

```python
sns.violinplot(x='tip',data=data,split='True')
```

**2. for numerical and categorical dataa**

```python
sns.violinplot(x='size',y='total_bill',split=True,data=data)
```

**hue parameter**

```python
sns.violinplot(x='day',y='tip',data=data,hue='sex')
```

**split function use to reduce cluttring(space utilization is optimized)**

**by using catplot function**

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',kind='violin')
```
