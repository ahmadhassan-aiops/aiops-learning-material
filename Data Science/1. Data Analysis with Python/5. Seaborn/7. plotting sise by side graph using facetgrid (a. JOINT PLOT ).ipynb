- A **FacetGrid** is a multi-axes grid with subplots visualizing the distribution of variables of a dataset and the relationship between multiple variables

- A **Jointplot** comprises three plots. Out of the three, one plot displays a bivariate graph which shows how the dependent variable(Y) varies with the independent variable(X). Another plot is placed horizontally at the top of the bivariate graph and it shows the distribution of the independent variable(X). The third plot is placed on the right margin of the bivariate graph with the orientation set to vertical and it shows the distribution of the dependent variable(Y). **It is very helpful to have univariate and bivariate plots together in one figure**. This is because the univariate analysis focuses on one variable, it describes, summarizes and shows any patterns in your data and the bivariate analysis explores the relationship between two variables and also describes the strength of their relationship. The function jointplot() in Seaborn library creates a scatter plot with two histograms at the top and right margins of the graph by default.

CAN BE USED IN FOUR MANNER:
- a.	scatter plot
- b.	Hexbinplot
- c.	Kdeplot
- d.	Regplot
- **ALWAYS ON NUMERICAL DATA**


```python
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=sns.load_dataset("tips")
data.head()
```

#A. SCATTER PLOT

- By default joinplot function buids the scatter plot.

```python
sns.jointplot(x='total_bill',y='tip',data=data)
```

#b. hexplot (2D histogram)

- **the darker shade represent higher frequency and viceversa**
- It is preffered when we want to see the conaentrated area of data.

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='hex')
```

#c.	KDE plot (kemel density estimation) or PDF

A contour plot (2D kde plot ) is a graphical technique for representing a 3-dimensional surface by plotting constant z slices, called contours, on a 2-dimensional format. That is, given a value for z, lines are drawn for connecting the (x,y) coordinates where that z value occurs


```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde',color='black')
```

# Reg plot

Regression plots as the name suggests creates a regression line between 2 parameters and helps to visualize their linear relationships It produces "best fit line". Thus used for the estimation of future values.


```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```
