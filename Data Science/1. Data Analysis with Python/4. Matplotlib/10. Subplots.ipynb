#Stateful vs stateless in matplotlib


1. **Stateful (Pyplot)**: Pyplot handles the figure and axes for you, keeping track of the current plot. It's useful for quick, simple plots because you don't need to manage everything yourselfmatplotlib does it behind the scenes.

2. **Stateless**: A stateless function doesnt remember anything from before. It only depends on the data you give it at that moment, not anything outside or from the past. Each time it runs, it works as if it's starting fresh.

3. **Figure and Axes**:
   - **Figure (fig)**: A container that can hold one or more plots.
   - **Axes**: The individual plots inside the figure.

In short we can say that, **fig** can have multiple plots (axes), and Pyplot automatically manages these for you in a stateful way, while stateless functions need everything explicitly given each time.

#TO DRAW SEPERATE PLOTS BUT ON SAME FIG

- SHAREX ==> TO RETAIN X AXIS LABELS ON LOWER MOST
- SHAREY ==> TO RETAIN Y AXIS LABELS ON LOWER MOST
- NO NEED OF X AXIS ON 1 AND 2

```python
plt.figure(figsize=(15,10))

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig,(axis1,axis2,axis3)=plt.subplots(nrows=3,ncols=1,sharex=True)

axis1.plot(ages, py_salaries, label='Python')
axis2.plot(ages, js_salaries, label='JavaScript')
axis3.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')


axis1.legend()
axis1.set_title('Median Salary (USD) by Age')
axis1.set_ylabel('Median Salary (USD)')

axis2.legend()
axis2.set_title('Median Salary (USD) by Age')
axis2.set_ylabel('Median Salary (USD)')

axis3.legend()
axis3.set_title('Median Salary (USD) by Age')
axis3.set_xlabel('Ages')
axis3.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

#TO DRAW SEPERATE PLOTS ON DIFFERENT FIG

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/1. Roadmap for EDA/4.MATPLOTLIB/data subplot.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1,axis1=plt.subplots()
fig2,axis2=plt.subplots()
fig3,axis3=plt.subplots()

axis1.plot(ages, py_salaries, label='Python')
axis2.plot(ages, js_salaries, label='JavaScript')
axis3.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')


axis1.legend()
axis1.set_title('Median Salary (USD) by Age')
axis1.set_ylabel('Median Salary (USD)')
axis1.set_xlabel('Ages')

axis2.legend()
axis2.set_title('Median Salary (USD) by Age')
axis2.set_ylabel('Median Salary (USD)')
axis2.set_xlabel('Ages')

axis3.legend()
axis3.set_title('Median Salary (USD) by Age')
axis3.set_xlabel('Ages')
axis3.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
