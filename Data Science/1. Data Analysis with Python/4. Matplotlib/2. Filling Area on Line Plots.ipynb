##1.Filling between ages and python salaries

```python
ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median =57287
plt.fill_between(ages, py_salaries,overall_median,alpha=0.7)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

##3.Conditional filling between ages,python salaries and overall median salaries

**Interpolation** means determining a value from the existing values in a given data set. Another way of describing it is the act of inserting or interjecting an intermediate value between two other values

The interpolate parameter in plt.fill_between() determines whether to interpolate the fill region where there are gaps in the data. This parameter becomes relevant when the where condition introduces discontinuities in the filled region.



```python
ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median =57287
plt.fill_between(ages, py_salaries,overall_median,alpha=0.5,where=(overall_median>py_salaries),interpolate=False,label='lower than median salaries')
plt.fill_between(ages, py_salaries,overall_median,alpha=0.5,where=(overall_median<=py_salaries),interpolate=False,label='above than median salaries')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

##4.Conditional filling between ages,python salaries and all developers salaries

```python
ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='js_salaries')
plt.plot(ages, py_salaries, label='Python')

overall_median =57287
plt.fill_between(ages, py_salaries,js_salaries,alpha=0.75,where=(js_salaries>py_salaries),interpolate=True,label='Above Avg')
plt.fill_between(ages, py_salaries,js_salaries,alpha=0.25,where=(js_salaries<=py_salaries),interpolate=True,label='below Avg')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```
