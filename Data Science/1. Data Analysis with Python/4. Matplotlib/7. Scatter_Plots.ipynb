#What is a scatter plot?

- Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensionalplane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis.

- A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship betweentwo variables, for example by looking for any correlation between them.

#When to use it?

- When we are interested to know the relationship between two numerical variables.

```python
import matplotlib as mp
from matplotlib import pyplot as plt
import pandas as pd
```

#Basic scatter plot

- Random values in our two variables thus scatter plot confirms there is no correlation.

      as **sizes** are small thus we customized the sizes(s)
      for customized **colors**(c="")
      for customized **markers**(markers="")
      for **edgecolor**(around marker)
      for **linewidth**(of edgecolor)
      for **alpha**(used color shade)


```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x,y,s=30,c='blue',marker=".",edgecolor="black",linewidth=7,alpha=0.9)

plt.tight_layout()
```

#To add colors using cmap

- Here we add colors in the form of cmap which represent the response quality i.e light color closer to 0's while darker are closer to 10's

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y,s=100,c=colors,marker="v",cmap="Blues",edgecolor="black",linewidth=1,alpha=1)
plt.tight_layout()
```

#COLOR BARS with different SIZES

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [50, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y,c=colors,marker="v",cmap="Blues",edgecolor="black",linewidth=1,alpha=1,s=sizes)
plt.tight_layout()
```

#REAL WORLD DATA EXAMPLE


```python
from matplotlib.figure import cbar
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CLIENTS/12. Sarah/Matplotlib/data scatter plot.csv')
data.head()
```

```python
data.shape
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,edgecolor='black',linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
```

#Adding log values and cmap color with colorbar

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,edgecolor='black',c=ratio, linewidth=1, alpha=0.75,cmap='summer')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
```
