```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from datetime import datetime,timedelta
from matplotlib import dates as mpl_dates
```

#BASIC TIME SERIES PLOT

- BY DEFAULT IT HAS NO LINE SO WE USE ARGUEMENT OF 'LINESTYLE'
- MAKERS ARE PRESENT BBY DEFAULT,CAN BE CHANGED BY MARKER=NONE

```python
dates = [datetime(2019, 5, 24),
         datetime(2019, 5, 25),
         datetime(2019, 5, 26),
         datetime(2019, 5, 27),
         datetime(2019, 5, 28),
         datetime(2019, 5, 29),
         datetime(2019, 5, 30)]
y = [1,2, 3, 4, 6, 5, 7]

plt.plot_date(dates,y,linestyle="solid")

plt.tight_layout()
plt.style.use("fivethirtyeight")
```

#FORMATTING OF FIGURE(date style)

- this method only works on figure and not on pyplot gcf == get current figure

```python
dates = [datetime(2019, 5, 24),
         datetime(2019, 5, 25),
         datetime(2019, 5, 26),
         datetime(2019, 5, 27),
         datetime(2019, 5, 28),
         datetime(2019, 5, 29),
         datetime(2019, 5, 30)]
y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y,linestyle="solid")

plt.tight_layout()
plt.style.use("fivethirtyeight")

plt.gcf().autofmt_xdate()
```

#DATE FORMAT

```python
dates = [datetime(2019, 5, 24),
         datetime(2019, 5, 25),
         datetime(2019, 5, 26),
         datetime(2019, 5, 27),
         datetime(2019, 5, 28),
         datetime(2019, 5, 29),
         datetime(2019, 5, 30)]
y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y,linestyle="solid")
plt.tight_layout()
plt.style.use("fivethirtyeight")
plt.gcf().autofmt_xdate()

date_format=mpl_dates.DateFormatter('%d %b, %y')
plt.gca().xaxis.set_major_formatter(date_format)
```

#REAL WORLD EXAMPLE(BITCOIN DATA)

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CLIENTS/12. Sarah/Matplotlib/data TIME SERIES.csv')
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(12,4))

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.tight_layout()
plt.style.use("fivethirtyeight")
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price of bitcoin')

date_format=mpl_dates.DateFormatter('%d %b, %y')
plt.gca().xaxis.set_major_formatter(date_format)
```

#THE DATES PRESENT AS STRINGS CONVERTED INTO DATESTIME FORMAT!

```python
#changing data dytpe of date field into datetime format for further processing
data['Date'] = pd.to_datetime(price_date)
data.sort_values('Date', inplace=True)
```

```python
# to confirm datatype of date field
data.info()
```

```python
price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.tight_layout()
plt.style.use("fivethirtyeight")

date_format=mpl_dates.DateFormatter('%b %d, %y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```
