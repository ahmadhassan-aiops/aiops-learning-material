```python
import matplotlib as mp
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

#Pie chart and its use

- IT is strictly used for the representation of your categories (To represent proportion of your categorical data)
- It is best practice to represent the categories in pie chart range from 5 to 10.

```python
values=(56,55,58)

plt.pie(values)

plt.tight_layout()
plt.title("pie chart")
plt.style.use("fivethirtyeight")
```

##LABELS

```python
plt.tight_layout()

plt.title("pie chart")
plt.style.use("fivethirtyeight")

values=(30,55,70)
labels=["thirty","fifty five","seventy"]

plt.pie(values,labels=labels)
plt.show
```

##WEDGE

```python
values=(30,55,55)
plt.pie(values,labels=labels,wedgeprops={'edgecolor':"black"})

labels=["thirty","fifty five","fifty five"]
plt.tight_layout()
plt.title("pie chart")
plt.style.use("fivethirtyeight")
```

##COLORS (built in) can be called by names

```python
plt.tight_layout()
plt.title("pie chart")
plt.style.use("fivethirtyeight")

values=(30,55,55)
labels=["thirty","fifty five","fifty five"]

colors=["cyan","yellow","green"]

plt.pie(values,labels=labels,colors=colors,wedgeprops={'edgecolor':"black"})
plt.show()
```

##hex color values

```python
plt.tight_layout()
plt.title("pie chart")
plt.style.use("fivethirtyeight")
values=(30,55,55)
labels=["thirty","fifty five","fifty five"]

colors=['#008fd5','#fc4f30','#e5ae37']

plt.pie(values,labels=labels,colors=colors,wedgeprops={'edgecolor':"black"})
plt.show()
```

##REAL WORLD EXAMPLE

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920,333,444,777777]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Ot',"as","xjh",'react']

plt.pie(responses,labels=programming_language,wedgeprops={'edgecolor':"black"})
plt.show()
```

**IF WE HAVE MORE THAN 5 ITEMS THEN BAR CHARTS ARE MORE REPRESENTABLE LIKE IN THIS CASE ITEMS GETS ITERMIXED WITH EACHOTHER**

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(responses,labels=programming_language,wedgeprops={'edgecolor':"black"})
plt.show()
```

## HOW TO EMPHASIZE ON ONE VALUE (explode)?

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode=[0,0,0,0.2,0]

plt.pie(responses,labels=programming_language,explode=explode,wedgeprops={'edgecolor':"black"})
plt.show()
```

##SHADOW

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0.1,0,0,0,0]

plt.pie(responses,labels=programming_language,explode=explode,shadow=True,wedgeprops={'edgecolor':"black"})
plt.show()
```

##START ANGLE

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode=[0,0,0,0.1,0]
plt.pie(responses,labels=programming_language,explode=explode,startangle=180,shadow=True,wedgeprops={'edgecolor':"black"})
plt.show()
```

##PERCENTAGES

```python
plt.tight_layout()
plt.title("POPULARITY OF PROGRAMMING LANGUAGE")
plt.style.use("fivethirtyeight")

responses = [59219, 55466, 47544, 36443, 35917]
programming_language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode=[0,0,0,0,0.1]
plt.pie(responses,labels=programming_language,explode=explode,startangle=90,autopct="%1.1f%%",shadow=True,wedgeprops={'edgecolor':"black"})
plt.show()
```
